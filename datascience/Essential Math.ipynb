## Chain rule

```python
x,y,z = symbols("x y z")
z = (x**2 + 1)**3 - 2
dz_dx = diff(z, x)
print(dz_dx)
```

$$ 
y = x^2 + 1 \\
z = y^3 − 2  \\
z= (x^2+1)^3 -2 \\
\frac{dy}{dx} (x^2 + 1) = 2x \\ 
\frac{dz}{dy} (y^3 − 2)= 3y^2 \\
\frac{dz}{dx} = 2x * 3y^2 \\
\frac{dz}{dx} = 2x * 3(x^2 +1)^2 \\
$$



$$
\begin{array}{ll}
\textbf{Définitions} & \textbf{Dérivées} \\
y = x^2 + 1 & \frac{dy}{dx}(x^2 + 1) = 2x \\
z = y^3 - 2 & \frac{dz}{dy}(y^3 - 2) = 3y^2 \\
z = (x^2 + 1)^3 - 2 & \frac{dz}{dx} = 2x \cdot 3y^2 \\
& \frac{dz}{dx} = 2x \cdot 3(x^2 + 1)^2 \\
\end{array}
$$


```python
# need to underscore y to prevent variable clash
_y = x**2 + 1
dy_dx = diff(_y)
# derivative for second function
z = y**3 - 2
dz_dy = diff(z)
# Calculate derivative with and without
# chain rule, substitute y function
dz_dx_chain = (dy_dx * dz_dy).subs(y, _y)
dz_dx_no_chain = diff(z.subs(y, _y))
# Prove chain rule by showing both are equal
print(dz_dx_chain) # 6*x*(x**2 + 1)**2
print(dz_dx_no_chain) # 6*x*(x**2 + 1)**2
```

## Integrals

```python
def approximate_integral(a, b, n, f):
    delta_x = (b - a) / n
    total_sum = 0
    for i in range(1, n + 1):
        midpoint = 0.5 * (2 * a + delta_x * (2 * i - 1))
        total_sum += f(midpoint)
    return total_sum * delta_x
def my_function(x):
    return x**2 + 1
area = approximate_integral(a=0, b=1, n=50, f=my_function)
print(area)
```

# Chapter 2 - Probability

Probability is about quantifying predictions of events yet to happen, whereas
likelihood is measuring the frequency of events that already occurred. In statistics
and machine learning, we often use likelihood (the past) in the form of data to
predict probability (the future).

La probabilité consiste à quantifier les prédictions d’événements qui ne se sont pas encore produits, tandis que la vraisemblance mesure la fréquence des événements déjà survenus. En statistique et en apprentissage automatique, on utilise souvent la vraisemblance (le passé), sous forme de données, pour prédire la probabilité (le futur)

Alternatively, probability can be expressed as an odds O(X) such as 7:3, 7/3, or 2.333.
To turn an odds O(X) into a proportional probability P(X) , use this formula

Autrement, la probabilité peut être exprimée sous forme de cote (odds) $O_X$, comme 7:3, 7/3 ou 2,333.
Pour convertir une cote $O_X$ en une probabilité proportionnelle $P_X$, utilisez la formule suivante 

$ P(x) = \frac {O(X)}{1+O(X)} $ 

Exemple : une cote de 3 contre 1 (3 chance que l'évènement se produise, 1 qu'il ne se produise pas) a une probabilité de $ \frac {3}{3+1} = 0,75  $



Une cote de **7:5** (au sens probabiliste, *odds*) signifie :

* **7 chances favorables** (l’événement se produit)
* **5 chances défavorables** (l’événement ne se produit pas)

Total des cas considérés = $7 + 5 = 12$.

👉 Probabilité que l’événement se produise :

$$
P = \frac{7}{7+5} = \frac{7}{12}
$$

$$
P \approx 0,5833 = 58,3\%
$$

👉 Probabilité qu’il ne se produise pas :

$$
1 - P = \frac{5}{12} \approx 41,7\%
$$

### Median

outliers : valeurs aberrantes

The median is actually the 50% quantile

```python
# Number of pets each person owns
def median(values):
    ordered = sorted(values)
    print(ordered)
    n = len(ordered)
    print(f"n%2 : {n%2}")
    print(f"n/2 : {int(n/2)}")
    mid = int(n / 2) - 1 if n % 2 == 0 else int(n/2)
    if n % 2 == 0:
        return (ordered[mid] + ordered[mid+1]) / 2.0
    else:
        return ordered[mid]
sample1 = [0, 1, 5, 20, 40, 50, 100]
print(median(sample1)) # prints 20
sample2 = [18,0, 1, 5, 6, 9, 10, 14]
print(median(sample2)) # prints 7
```

### Mode

```python
# Number of pets each person owns
from collections import defaultdict
sample = [1, 3, 2, 7, 7, 0, 2, 3]
def mode(values):
    counts = defaultdict(lambda: 0) # dictionnaire avec valeur par défaut = 0
    for s in values:
        counts[s] += 1 # incrémente la valeur de chaque compteur
    max_count = max(counts.values())
    print(max_count)
    modes = [v for v in set(values) if counts[v] == max_count]
    return modes
print(mode(sample)) # [2, 3]
```

### Variance & Standard deviation

écart type en Français

$$
\text{population variance} = \frac{(x_1 - \text{mean})^2 + (x_2 - \text{mean})^2 + \cdots + (x_n - \text{mean})^2}{N}
$$
$$
\sigma^2 = \frac{\sum (x_i - \mu)^2}{N}
$$



```python
# Number of pets each person owns
data = [0, 1, 5, 7, 9, 10, 14]
def variance(values):
    mean = sum(values) / len(values)
    _variance = sum((v - mean) ** 2 for v in values) / len(values)
    return _variance
print(variance(data)) # prints 21.387755102040813
```

The opposite of a square is a square root, so let’s take the square root of the variance,
which gives us the standard deviation.

$$
\sigma = \sqrt{ \frac{ \sum (x_i - \mu)^2 }{N} }
$$

sigma = écart type de la population et s écart type de l'échantillon


```python
def std_dev(values):
    return sqrt(variance(values))
print(std_dev(data)) # prints 4.624689730353898
```

there is an important tweak we need to apply to these two formulas
when we calculate for a sample, to decrease any bias in a sample and not underestimate the variance of the population based on our sample. By counting values short of one item in our divisor, we increase the variance and therefore capture greater uncertainty in our sample.

$$
s^2 = \frac{ \sum (x_i - \bar{x})^2 }{n - 1}
$$

$$
s = \sqrt{ \frac{ \sum (x_i - \bar{x})^2 }{n - 1} }
$$


```python
# Number of pets each person owns
data = [0, 1, 5, 7, 9, 10, 14]
def variance(values, is_sample: bool = False):
    mean = sum(values) / len(values)
    _variance = sum((v - mean) ** 2 for v in values) / (len(values) - (1 if is_sample else 0))
    return _variance
def std_dev(values, is_sample: bool = False):
    return sqrt(variance(values, is_sample))
print("VARIANCE = {}".format(variance(data, is_sample=True))) # 24.95238095238095
print("STD DEV = {}".format(std_dev(data, is_sample=True))) # 4.99523582550223
```

### Normal Distribution

The probability density function (PDF) that creates the normal distribution is as
follows:
$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} \, e^{- \frac{1}{2} \left( \frac{x - \mu}{\sigma} \right)^2}
$$


```python
# normal distribution, returns likelihood
def normal_pdf(x: float, mean: float, std_dev: float) -> float:
    return (1.0 / (2.0 * math.pi * std_dev ** 2) ** 0.5) * math.exp(-1.0 * ((x - mean) ** 2 / (2.0 * std_dev ** 2)))
```

1. The mean of the sample means is equal to the population mean.
2. If the population is normal, then the sample means will be normal.
3. If the population is not normal, but the sample size is greater than 30, the sample
means will still roughly form a normal distribution.
4. The standard deviation of the sample means equals the population standard
deviation divided by the square root of n:

- La moyenne des moyennes d’échantillons est égale à la moyenne de la population.
- Si la population est normale, alors les moyennes d’échantillons seront normales.
- Si la population n’est pas normale, mais que la taille de l’échantillon est supérieure à 30, les moyennes d’échantillons formeront tout de même approximativement une distribution normale.
- L’écart‑type des moyennes d’échantillons est égal à l’écart‑type de la population divisé par la racine carrée de $n$ :


$$ \text{sample standard deviation} = \frac { \text {population standard deviation}}
{\sqrt \text{sample size}} $$

### Confidence Interval

« Vous avez peut‑être déjà entendu parler du terme « intervalle de confiance », qui embrouille souvent les débutants et les étudiants en statistique. Un intervalle de confiance est un calcul sous forme de plage de valeurs indiquant à quel point nous sommes confiants que la moyenne d’un échantillon (ou un autre paramètre) se situe dans un certain intervalle pour la moyenne de la population. »


Based on a sample of 31 golden retrievers with a sample mean of 64.408 and a sample
standard deviation of 2.05, I am 95% confident that the population mean lies between
63.686 and 65.1296. How do I know this? Let me show you, and if you get confused,
circle back to this paragraph and remember what we are trying to achieve. I highlighted
it for a reason!

level of confidence (LOC) : I want to be 95% confident that my
sample mean falls in the population mean range I will calculate

First, I need the critical z-value which is the symmetrical range in a standard
normal distribution that gives me 95% probability in the center

```python
from scipy.stats import norm
def critical_z_value(p):
    norm_dist = norm(loc=0.0, scale=1.0)
    left_tail_area = (1.0 - p) / 2.0
    upper_area = 1.0 - ((1.0 - p) / 2.0)
    return norm_dist.ppf(left_tail_area), norm_dist.ppf(upper_area)
print(critical_z_value(p=.95))
# (-1.959963984540054, 1.959963984540054)
```

Next I’m going to leverage the
central limit theorem to produce the margin of error (E),

$$
E = \pm z_{c} \frac{s}{\sqrt{n}} \\
E = \pm 1.95 \frac{2.05}{\sqrt{31}}  \\
E = 0.72164
$$


If we apply that margin of error against the sample mean, we finally get the confidence
interval

95% confidence interval = 64.408 ± 0.72164

```python
def confidence_interval(p, sample_mean, sample_std, n):
    # Sample size must be greater than 30
    lower, upper = critical_z_value(p)
    lower_ci = float(lower * (sample_std / sqrt(n)))
    upper_ci = float(upper * (sample_std / sqrt(n)))
    return sample_mean + lower_ci, sample_mean + upper_ci

print(confidence_interval(p=.95, sample_mean=64.408, sample_std=2.05, n=31))
```

So the way to interpret this is “based on my sample of 31 golden retriever weights
with sample mean 64.408 and sample standard deviation of 2.05, I am 95% confident
the population mean lies between 63.686 and 65.1296.” That is how we describe our
confidence interval
