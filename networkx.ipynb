```python
import pandas as pd
import networkx as nx

# Importer les dépendances de Pandas
dependencies = pd.__file__.split("/")[1:-1]

# Créer un graphe
graph = nx.Graph()

# Ajouter les nœuds
for dependency in dependencies:
    graph.add_node(dependency)

# Ajouter les liens
for i in range(len(dependencies)):
    for j in range(i + 1, len(dependencies)):
        if dependencies[i] in pd.__file__.split("/")[dependencies.index(dependency) + 1:dependencies.index(dependency) + 1 + dependencies.index(dependencies[j])]:
            graph.add_edge(dependencies[i], dependencies[j])

# Filtrer les dépendances utilisées
used_dependencies = []
for dependency in graph.nodes:
    if dependency in ['pandas', 'numpy', 'typing', 'setuptools', 'wheel', 'packaging', 'pytest', 'pytest-runner']:
        used_dependencies.append(dependency)

# Afficher le graphe
nx.draw(graph.subgraph(used_dependencies), with_labels=True)
```
