## Build adjacent Matrix from the graph

```python
G = nx.Graph()
L1 = [(0,1),(1,2),(0,6),(5,6),(1,6),(2,5),(2,3),(2,4),(2,6),(1,7)]
G.add_edges_from(L1)

# Obtenir la matrice d'adjacence
adj_matrix = nx.adjacency_matrix(G)

# Convertir la matrice d'adjacence en une matrice numpy
adj_matrix_np = adj_matrix.toarray()


print("Matrice d'adjacence :")
print(adj_matrix_np)
```

Cependant, il y a une nuance subtile entre todense() et toarray(). La mthode toarray() est une mthode gnrique que vous pouvez utiliser pour convertir divers types de matrices (sparse ou dense) en une matrice NumPy dense. En revanche, todense() est spcifique aux matrices creuses (sparse) et est une mthode de la classe scipy.sparse qui est utilise pour reprsenter des matrices creuses.

En pratique, si vous savez que votre matrice est dense (c'est gnralement le cas pour la matrice d'adjacence complte d'un graphe), vous pouvez utiliser l'une ou l'autre sans problme. Cependant, si votre matrice est potentiellement creuse, vous pouvez utiliser todense() pour tre spcifique  la reprsentation sparse.

```python
adj_matrix_nx = nx.adjacency_matrix(G).todense()

print("Matrice d'adjacence :")
print(adj_matrix_nx)
```

## Eigenvalues and eigenvectors

```python
from numpy.linalg import eigvals, eig

# Create a graph
G = nx.Graph()
edges = [(0, 1), (1, 2), (0, 6), (5, 6), (1, 6), (2, 5), (2, 3), (2, 4), (2, 6), (1, 7)]
G.add_edges_from(edges)

# Get the adjacency matrix
adj_matrix = nx.adjacency_matrix(G).todense()

# Find the eigenvalues
eigenvalues = eigvals(adj_matrix)

# Find the eigenvectors
eigenvalues, eigenvectors = eig(adj_matrix)

print("Eigenvalues:")
print(eigenvalues)

print("\nEigenvectors:")
print(eigenvectors)
```
